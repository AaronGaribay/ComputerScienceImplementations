### SEQUENCE LABELING WITH FUNSD

The autors of LayoutLM [provides a fine-tuning](https://github.com/microsoft/unilm/tree/master/layoutlm#sequence-labeling-task) implementation with the Funsd dataset as part of the sequency labeling task that this pre-train provides, so here is the implementation.

First we download the datazet, unziped, move dataset to a new folder 'data'.

Optionaly we delete dataset.zip

Here comes the most important part:

The FUNDS datast comes with its own labelling method, its described in its [paper](https://arxiv.org/pdf/1905.13538.pdf). So the funsd_preproccess.py is a script that extracts the labels from the trainig data .json annotations with the BertTokenizer transformation
